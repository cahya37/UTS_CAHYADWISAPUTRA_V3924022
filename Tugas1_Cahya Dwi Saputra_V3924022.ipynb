```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Set style visualisasi
sns.set(style="whitegrid")

# Parameter distribusi Poisson
lam = 8  # rata-rata pelanggan per jam
jam_operasional = 12
jumlah_hari = 30
total_jam = jam_operasional * jumlah_hari

# Simulasi kedatangan pelanggan
kedatangan = np.random.poisson(lam=lam, size=total_jam)

# Buat DataFrame
df_kedatangan = pd.DataFrame({'Jumlah Pelanggan': kedatangan})

# Hitung probabilitas tidak ada pelanggan dalam satu jam
print("\nA:")
prob_tidak_ada = np.mean(kedatangan == 0)
print(f"Probabilitas tidak ada pelanggan dalam satu jam: {prob_tidak_ada:.2f}")

# Hitung probabilitas lebih dari 10 pelanggan dalam satu jam
print("\nB:")
prob_lebih_dari_10 = np.mean(kedatangan > 10)
print(f"Probabilitas lebih dari 10 pelanggan dalam satu jam: {prob_lebih_dari_10:.2f}")

# Statistik deskriptif
print("\nC:")
print("\nStatistik Deskriptif:")
print(df_kedatangan.describe())

# Visualisasi histogram dan distribusi teoritis
plt.figure(figsize=(10, 6))
sns.histplot(kedatangan, bins=range(0, max(kedatangan)+1), kde=False, stat="density", color='skyblue', label='Simulasi')
x_vals = np.arange(0, max(kedatangan)+1)
plt.plot(x_vals, poisson.pmf(x_vals, mu=lam), 'r--', label='Distribusi Teoritis (Poisson)', linewidth=2)
plt.title("Distribusi Kedatangan Pelanggan per Jam (Simulasi vs Teoritis)")
plt.xlabel("Jumlah Pelanggan per Jam")
plt.ylabel("Frekuensi (Density)")
plt.legend()
plt.tight_layout()
plt.show()
```
